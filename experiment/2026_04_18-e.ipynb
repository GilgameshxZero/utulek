```python
def diff(X: float, N_max: int):
	lhs_base = np.zeros((N_max - 1, )) + 1 + X
	lhs_exp = np.arange(2, N_max + 1)
	lhs = np.power(lhs_base, lhs_exp)
	rhs = 1 + np.arange(2, N_max + 1) * X
	return lhs - rhs
```
