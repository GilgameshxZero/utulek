```python
def sim(n: int, switch: bool):
	labels = np.random.randint(3, size=n)
	preds = np.random.randint(3, size=n)
	reveals = np.random.randint(3, size=n)
	reveals = (reveals + (reveals == labels)) % 3
	reveals = (reveals + (reveals == preds)) % 3
	reveals = (reveals + (reveals == labels)) % 3
	if switch:
		preds = 3 - preds - reveals
	score = np.sum(preds == labels, axis=0) / n
	return score
```

```python
print("No switch:", sim(65536, False))
print("Switch:", sim(65536, True))
```
