```python
import re
import os

pattern = re.compile(r"(\d\d\d\d)-(\d\d)-(\d\d)(.*)")

for i in glob.glob("**/*"):
    filename_len = len(i.split("/")[-1])
    pathname = i[:-filename_len]
    filename = i[-filename_len:]
    match = pattern.fullmatch(filename)
    if not match:
        continue
    groups = match.groups()
    suffix = groups[3].replace("-", "_")
    if suffix[0] == "_":
        suffix = "-" + suffix[1:]
    new_name = f"{groups[0]}_{groups[1]}_{groups[2]}{suffix}"
    print(pathname + filename, pathname + new_name)
    # os.rename(pathname + filename, pathname + new_name)
```

```python
import re
import os

pattern = re.compile(r"(\d\d\d\d)-(\d\d)-(\d\d)(.*)")

for i in glob.glob("../altair/*/*/*.svg"):
    filename_len = len(i.split("/")[-1])
    pathname = i[:-filename_len]
    filename = i[-filename_len:]
    print(filename)
    match = pattern.fullmatch(filename)
    if not match:
        continue
    groups = match.groups()
    suffix = groups[3].replace("-", "_")
    if suffix[0] == "_":
        suffix = "-" + suffix[1:]
    new_name = f"{groups[0]}_{groups[1]}_{groups[2]}{suffix}"
    print(pathname + filename, pathname + new_name)
    os.rename(pathname + filename, pathname + new_name)
```
