```python
pattern = re.compile(r"(\d\d\d\d)-(\d\d)-(\d\d) (.+)")

for i in glob.glob("../../cosine/archive.syncthing/*-*-* *"):
    filename_len = len(i.split("/")[-1])
    pathname = i[:-filename_len]
    filename = i[-filename_len:]
    match = pattern.fullmatch(filename)
    if not match:
        continue
    groups = match.groups()
    new_name = f"{groups[0]}_{groups[1]}_{groups[2]}-{groups[3].replace(' ', '_')}"
    print(pathname + filename, pathname + new_name)
    os.rename(pathname + filename, pathname + new_name)
```
