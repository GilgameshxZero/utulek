```python
import utulek

dss = (utulek.fetch_tfds_img_ds("mnist", "train[:1024]"),
       utulek.fetch_tfds_img_ds("mnist", "test[:256]"))
svc = utulek.train_svc(1, *dss[0], kernel="linear")
acc = utulek.eval_svc_acc(svc, *dss[1])
print(acc)
pred = svc.predict(utulek.transform_flatten_img_ds_X(dss[1][0]))
```

```python
utulek.show_img_ds(dss[1][0], dss[1][1], pred, shape=(8, 8))
```

```python
dss = (utulek.fetch_tfds_img_ds("cifar10", "train[:1024]"),
       utulek.fetch_tfds_img_ds("cifar10", "test[:256]"))
labels = utulek.fetch_tfds_img_ds_labels("cifar10")
svc = utulek.train_svc(1, *dss[0], kernel="linear")
acc = utulek.eval_svc_acc(svc, *dss[1])
print(acc)
pred = svc.predict(utulek.transform_flatten_img_ds_X(dss[1][0]))
utulek.show_img_ds(dss[1][0], dss[1][1], pred, labels, shape=(8, 8))
```
