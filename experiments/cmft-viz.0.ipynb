```python
# ! rm -rf .data/cmft-viz.ipynb
! mkdir .data
! mkdir .data/cmft-viz.ipynb

import dill
import matplotlib.pyplot as plt
import numpy as np

res = []
with open(".data/cmft-eval.ipynb/1_1.dill", "rb") as f:
    res.append(dill.load(f))
with open(".data/cmft-eval.ipynb/1_2.dill", "rb") as f:
    res.append(dill.load(f))
with open(".data/cmft-eval.ipynb/1_3.dill", "rb") as f:
    res.append(dill.load(f))
with open(".data/cmft-eval.ipynb/1_4.dill", "rb") as f:
    res.append(dill.load(f))
with open(".data/cmft-eval.ipynb/1_3.dill", "rb") as f:
    res.append(dill.load(f))
with open(".data/cmft-eval.ipynb/2_1.dill", "rb") as f:
    res.append(dill.load(f))
with open(".data/cmft-eval.ipynb/1_7.dill", "rb") as f:
    res.append(dill.load(f))

res[2] = res[1][1:] + res[2][:2]

[print(r) for r in res];
```

```python
data = {
    "G35": res[0][0],
    "W53 P1 Plain": res[0][1],
    "W53 P2 Plain": res[0][2],
    "W53 P1": res[0][3],
    "W53 P2": res[0][4],
}

courses = list(data.keys())
values = list(data.values())
 
fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(courses, values, color ='black', 
        width = 0.4)
plt.ylim(0, 1)
plt.ylabel("Capabilities")
plt.show()
fig.savefig(".data/cmft-viz.ipynb/w2-1.png", transparent=True)
```
